```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("housing.csv")
```

```python
df['income_cat'] = pd.cut(df['median_income'], bins=[0, 1.5, 3.0, 4.5, 6.0, np.inf], labels=[1,2,3,4,5])
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np # move test set to a separate file
```

```python
# Iterate split to get data
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(df, df['income_cat']):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```

```python
# Lets remove the income_cat column
for sett in (strat_train_set, strat_test_set):
    sett.drop("income_cat", axis=1, inplace=True)
```

```python
strat_test_set
```

```python
# create a copy of training data to make changes to data and not impact trainingset
df = strat_test_set.copy()
```

```python
# scatter plot
df.plot(kind="scatter", x="latitude", y="longitude", grid=True, alpha=0.3)
```
